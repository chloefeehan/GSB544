1. Identify, and list here, what aesthetics are being used and which variables are being mapped to each aesthetic.


The aesthetics being used in this are the x and y variables (Income and Life expectantcy), size, and fill color. Income is being mapped to x, life expectancy is being mapped to y, population is mapped to size, and four regions is mapped to fill color. 

2. Import the q1data.csv dataset.


```python
import pandas as pd

#load in data
q1data = pd.read_csv("/Users/chloefeehan/Desktop/python/GSB544/Labs/Lab1/data/q1data.csv")
q1data = q1data.dropna()
```

3. Recreate the plot as best you can using plotnine.

```python
from plotnine import ggplot, geom_point, aes, labs, scale_size, scale_x_continuous, scale_y_continuous
from plotnine import theme, theme_bw, scale_x_log10, scale_fill_manual, element_text
from plotnine import guides, geom_text, scale_fill_discrete, scale_x_continuous, scale_y_log10


(ggplot(q1data,
aes(x = "income", y = "life_exp", size = "population", fill = "four_regions"))
+ geom_text(aes(x = 7000, y = 60, label="2010"), size=120, color = "lightgray")
+ geom_text(aes(x = 32000, y = 5), label = "per person (GDP/capita, PPPS inflation-adjusted)", 
                                    size=6, color = "lightgray")
+ geom_text(aes(x = 300, y = 90), label = "years", size = 6, color = "lightgray", angle = 90)
+ geom_point(alpha = 0.7, color = "black", stroke = 0.2)
+ labs(x= "Income", y = "Life expectancy", fill = "World Regions")
+ scale_size(range = [.1, 24])
+ scale_x_log10(limits=[0, 128000], 
                breaks=[500, 1000, 2000, 4000, 8000, 16000, 32000, 64000],
                labels=["500", "1000", "2000", "4000", "8000", "16K", "32K", "64K"])
+ scale_y_continuous(limits = (5, 90), breaks = range(10, 91, 10))
+ scale_fill_manual(values={'africa': 'aqua', 'americas': 'lime', 'asia': 'red', 'europe': 'yellow'},
                    labels = ["Africa","Americas","Asia","Europe"])
+ theme_bw()
+ theme(text=element_text(family='Comic Sans', size=12),
        axis_title=element_text(family='Comic Sans', size=11, color = "darkblue"),
        axis_text=element_text(family='Comic Sans', size=8, color = "gray"))
+ guides(size = "none")
#+ theme(legend_position="none")

)
```

4. What other geometries could you use here besides geom_point? Provide an actual plot of another geometry being used here and why it’s good or bad for visualizing these data.

```python
from plotnine import geom_tile

(ggplot(q1data,
aes(x = "income", y = "life_exp", size = "population", fill = "four_regions"))
+ geom_tile(aes(width=5000, height=5), color="black", alpha=0.7)
+ labs(x= "Income", y = "Life expectancy", fill = "World Regions")
+ scale_x_continuous(limits=[0, 128000], 
                    breaks=[500, 8000, 16000, 32000, 64000],
                    labels=["500", "8K", "16K", "32K", "64K"])
+ scale_y_continuous(limits = (5, 90), breaks = range(10, 91, 10))
+ scale_fill_manual(values={'africa': 'aqua', 'americas': 'lime', 'asia': 'red', 'europe': 'yellow'},
                    labels = ["Africa","Americas","Asia","Europe"])
+ theme_bw()
+ theme(text=element_text(family='Comic Sans', size=12),
        axis_title=element_text(family='Comic Sans', size=11, color = "darkblue"),
        axis_text=element_text(family='Comic Sans', size=8, color = "gray"))
+ guides(size = "none")
)
```

Another way to visualize this data is using the geom_tile function. However, this geometry is not as readable as geom_point, especially since the tiles are much larger than the points. In addition, the tile sizes are the same no matter the datapoint, so specifying for population size is not included in this graph.

3. Recreate the plot as best you can using plotnine.

```python

(ggplot(q2data,
aes(x = "exports", y = "imports", size = "energy", fill = "four_regions"))
+ geom_text(aes(x = 110, y = 200, label="1997"), size=120, color = "lightgray")
+ geom_point(alpha = 0.7, color = "black", stroke = 0.2)
+ labs(x= "Exports (/% of GDP)", y = "Imports (/% of GDP)", fill = "World Regions")
+ guides(size = "none")
+ scale_fill_manual(values={'africa': 'aqua', 'americas': 'lime', 'asia': 'red', 'europe': 'yellow'},
                    labels = ["Africa","Americas","Asia","Europe"])
+ scale_y_continuous(limits = (0, 400), breaks = range(50, 450, 50))
+ scale_x_continuous(limits = (0, 220), breaks = range(20, 220, 20))
+ theme_bw()
+ theme(text=element_text(family='Comic Sans', size=12),
        axis_title=element_text(family='Comic Sans', size=11, color = "darkblue"),
        axis_text=element_text(family='Comic Sans', size=8, color = "gray"))
)
```

4. What other geometries could you use here besides geom_point? Provide an actual plot of another geometry being used here and why it’s good or bad for visualizing these data.

```python
from plotnine import geom_violin 
(ggplot(q2data,
aes(x = "exports", y = "imports", fill = "four_regions"))
+ geom_violin(alpha = 0.3)
+ labs(x= "Exports (/% of GDP)", y = "Imports (/% of GDP)", fill = "World Regions")
+ guides(size = "none")
+ scale_fill_manual(values={'africa': 'aqua', 'americas': 'lime', 'asia': 'red', 'europe': 'yellow'},
                    labels = ["Africa","Americas","Asia","Europe"])
+ scale_y_continuous(limits = (0, 400), breaks = range(50, 450, 50))
+ scale_x_continuous(limits = (0, 220), breaks = range(20, 220, 20))
+ theme_bw()
+ theme(text=element_text(family='Comic Sans', size=12),
        axis_title=element_text(family='Comic Sans', size=11, color = "darkblue"),
        axis_text=element_text(family='Comic Sans', size=8, color = "gray"))
)
```

Another geometry that could be used to visualize this data is geom_violin. This would be a good way to look at the distributions of exports and imports if we waned to compare them. However, just by looking at this plot, we wouldn't want to use this since it overlaps the four regions a lot, making it hard to read. In addition, this visualization doesn't show energy which is a vital part of the visuzalition.
