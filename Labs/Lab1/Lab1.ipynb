## Task 1
1. Identify, and list here, what aesthetics are being used and which variables are being mapped to each aesthetic.

The aesthetics being used in this are the x and y variables (Income and Life expectantcy), size, and fill color. Income is being mapped to x, life expectancy is being mapped to y, population is mapped to size, and four regions is mapped to fill color. 

2. Import the q1data.csv dataset.


```python
import pandas as pd

#load in data
q1data = pd.read_csv("/Users/chloefeehan/Desktop/python/GSB544/Labs/Lab1/data/q1data.csv")
q1data = q1data.dropna()
```

3. Recreate the plot as best you can using plotnine.

```python
from plotnine import *


(ggplot(q1data,
aes(x = "income", y = "life_exp", size = "population", fill = "four_regions"))
+ geom_text(aes(x = 7000, y = 45, label="2010"), size=120, color = "lightgray")
+ geom_text(aes(x = 32000, y = 5), label = "per person (GDP/capita, PPPS inflation-adjusted)", 
                                    size=6, color = "lightgray")
+ geom_text(aes(x = 1000, y = 90), label = "INCOME LEVEL 1", 
                                    size=8, color = "lightgray")
+ geom_text(aes(x = 4000, y = 90), label = "LEVEL 2", 
                                    size=8, color = "lightgray") 
+ geom_text(aes(x = 12000, y = 90), label = "LEVEL 3", 
                                    size=8, color = "lightgray")    
+ geom_text(aes(x = 40000, y = 90), label = "LEVEL 4", 
                                    size=8, color = "lightgray")                               
+ geom_text(aes(x = 300, y = 90), label = "years", size = 6, color = "lightgray", angle = 90)
+ geom_point(alpha = 0.7, color = "black", stroke = 0.2)
+ labs(x= "Income", y = "Life expectancy", fill = "World Regions")
+ scale_size(range = [.1, 24])
+ scale_x_log10(limits=[0, 128000], 
                breaks=[500, 1000, 2000, 4000, 8000, 16000, 32000, 64000],
                labels=["500", "1000", "2000", "4000", "8000", "16K", "32K", "64K"])
+ scale_y_continuous(limits = (5, 90), breaks = range(10, 91, 10))
+ scale_fill_manual(values={'africa': 'aqua', 'americas': 'lime', 'asia': 'red', 'europe': 'yellow'},
                    labels = ["Africa","Americas","Asia","Europe"])
+ theme_bw()
+ theme(text=element_text(family='Comic Sans', size=12),
        axis_title=element_text(family='Comic Sans', size=11, color = "darkblue"),
        axis_text=element_text(family='Comic Sans', size=8, color = "gray"))
+ guides(size = "none")
+ theme(legend_position="none")

)
```

4. What other geometries could you use here besides geom_point? Provide an actual plot of another geometry being used here and why it’s good or bad for visualizing these data.

```python

(ggplot(q1data,
aes(x = "income", y = "life_exp", fill = "population"))
+ geom_tile(aes(width=5000, height=5), color="black", alpha=0.7, size = 0.3)
+ labs(x= "Income", y = "Life expectancy", fill = "Population")
+ scale_fill_gradient(low="lightblue", high="darkblue", limits=[0, 1e7])
+ scale_x_continuous(limits=[0, 128000], 
                    breaks=[500, 16000, 32000, 64000],
                    labels=["500", "16K", "32K", "64K"])
+ scale_y_continuous(limits = (5, 90), breaks = range(10, 91, 10))
+ facet_wrap("four_regions")
+ theme_bw()
+ theme(text=element_text(family='Comic Sans', size=12),
        axis_title=element_text(family='Comic Sans', size=11, color = "darkblue"),
        axis_text=element_text(family='Comic Sans', size=8, color = "gray"))
+ guides(size = "none")
)
```

Another way to visualize this data is using the geom_tile function. However, this geometry is not as readable as geom_point, especially since the tiles are much larger than the points. In addition, separating the four regions into different grids makes it harder to differentiate the different tiles, especially since they are so close together.

## Task 2

1. Identify, and list here, what aesthetics are being used and which variables are being mapped to each aesthetic.

The aesthetics being used in this are the x and y variables (Imports and Exports), size, and fill color. Exports is being mapped to x, Imports is being mapped to y, energy usage is mapped to size, and four regions is mapped to fill color. 

2. Import the q2data.csv dataset.


```python
q2data = pd.read_csv("/Users/chloefeehan/Desktop/python/GSB544/Labs/Lab1/data/q2data.csv")
q2data = q2data.dropna()
q2data["energy"] = q2data["energy"].str.replace("k", "").astype(float)
q2data.loc[q2data["energy"] == 12.40, "energy"] *= 1000
```

3. Recreate the plot as best you can using plotnine.

```python


(ggplot(q2data,
aes(x = "exports", y = "imports", size = "energy", fill = "four_regions"))
+ geom_text(aes(x = 110, y = 200, label="1997"), size=120, color = "lightgray")
+ geom_point(alpha = 0.7, color = "black", stroke = 0.2)
+ labs(x= "Exports (/% of GDP)", y = "Imports (/% of GDP)", fill = "World Regions")
+ guides(size = "none")
+ scale_fill_manual(values={'africa': 'aqua', 'americas': 'lime', 'asia': 'red', 'europe': 'yellow'},
                    labels = ["Africa","Americas","Asia","Europe"])
+ scale_size(range=[.1, 15])
+ scale_y_continuous(limits = (0, 400), breaks = range(50, 450, 50))
+ scale_x_continuous(limits = (0, 220), breaks = range(20, 220, 20))
+ theme_bw()
+ theme(text=element_text(family='Comic Sans', size=12),
        axis_title=element_text(family='Comic Sans', size=11, color = "darkblue"),
        axis_text=element_text(family='Comic Sans', size=8, color = "gray"))
+ theme(legend_position="none")
)
```

4. What other geometries could you use here besides geom_point? Provide an actual plot of another geometry being used here and why it’s good or bad for visualizing these data.

```python

(ggplot(q2data,
aes(x = "exports", y = "imports", fill = "four_regions"))
+ geom_violin(alpha = 0.3)
+ geom_point(aes(size = "energy"), alpha = 0.7, color = "black", position = "jitter")
+ labs(x= "Exports (/% of GDP)", y = "Imports (/% of GDP)", fill = "World Regions")
+ guides(size = "none")
+ facet_wrap("four_regions")
+ scale_fill_manual(values={'africa': 'aqua', 'americas': 'lime', 'asia': 'red', 'europe': 'yellow'},
                labels = ["Africa","Americas","Asia","Europe"])
+ scale_y_continuous(limits = (0, 150), breaks = range(50, 450, 50))
+ scale_x_continuous(limits = (0, 220), breaks = range(20, 220, 40))
+ theme_bw()
+ theme(text=element_text(family='Comic Sans', size=12),
        axis_title=element_text(family='Comic Sans', size=11, color = "darkblue"),
        axis_text=element_text(family='Comic Sans', size=8, color = "gray"))
)
```

Another geometry that could be used to visualize this data is geom_violin. This would be a good way to look at the distributions of exports and imports if we waned to compare them. However, just by looking at this plot, we wouldn't want to use this since the point sizes and violin densities overlap, making it hard to read. In addition, since this is broken up into 4 different areas, it makes it hard to compare the relationship between import, export, and energy all at once.

The aesthetics being used in this are the x and y variables, size, and fill color. internet_users is being mapped to x, GDP is being mapped to y, income is mapped to size, and four regions is mapped to fill color.

2. Import the q3data.csv dataset.


```python
q3data = pd.read_csv("/Users/chloefeehan/Desktop/python/GSB544/Labs/Lab1/data/q3data.csv")
q3data = q3data.dropna()
```

3. Recreate the plot as best you can using plotnine.

```python

(ggplot(q3data,
aes(x = "internet_users", y = "gdp", size = "income", fill = "four_regions"))
+ geom_text(aes(x = 50, y = 4500, label="2001"), size=120, color = "lightgray")
+ geom_text(aes(x = 90, y = 180), label = "\% of population", size=6, color = "lightgray")
+ geom_text(aes(x = 0, y = 40000), label = "USS inflation-adjusted", size=6, color = "lightgray", angle = 90)
+ geom_point(alpha = 0.8, color = "black", stroke = 0.2)
+ labs(x= "Individuals using the internet", y = "GDP/capita", fill = "World Regions")
+ scale_size(range = [.1, 14])
+ scale_x_continuous(limits = (0, 100), breaks = range(10, 100, 10))
+ scale_y_log10(limits = (0, 180000), 
                breaks=[200, 500, 1000, 2000, 5000, 10000, 20000, 50000, 100000],
                labels=["200", "500", "1000", "2000", "5000", "10K", "20K", "50K", "100K"])
+ guides(size = "none")
+ scale_fill_manual(values={'africa': 'aqua', 'americas': 'lime', 'asia': 'red', 'europe': 'yellow'},
                    labels = ["Africa","Americas","Asia","Europe"])
+ theme_bw()
+ theme(text=element_text(family='Comic Sans', size=12),
axis_title=element_text(family='Comic Sans', size=11, color = "darkblue"),
axis_text=element_text(family='Comic Sans', size=8, color = "gray"))
+ theme(legend_position="none")
)
```

4. What other geometries could you use here besides geom_point? Provide an actual plot of another geometry being used here and why it’s good or bad for visualizing these data.

```python
(ggplot(q3data,
aes(x = "internet_users", y = "gdp", size = "income", fill = "four_regions"))
+ geom_boxplot()
+ labs(x= "Individuals using the internet", y = "GDP/capita", fill = "World Regions")
+ scale_fill_manual(values={'africa': 'aqua', 'americas': 'lime', 'asia': 'red', 'europe': 'yellow'},
                    labels = ["Africa","Americas","Asia","Europe"])
+ theme_bw()
+ theme(text=element_text(family='Comic Sans', size=12),
                axis_title=element_text(family='Comic Sans', size=11, color = "darkblue"),
axis_text=element_text(family='Comic Sans', size=8, color = "gray"))
+ scale_x_continuous(limits = (0, 70), breaks = range(10, 100, 10))
+ scale_y_log10(limits = (0, 150000), 
                breaks=[200, 500, 1000, 2000, 5000, 10000, 20000, 50000, 100000],
                labels=["200", "500", "1000", "2000", "5000", "10K", "20K", "50K", "100K"])
)
```

Another way to visualize this plot would be the geom_boxplot function. However, since GDP and internet users are continuous variables, this would not be the most ideal visualization for the data. In addition, income is not represented in this dataset, which is an intergral part of the visualization.
