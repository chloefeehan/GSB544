The aesthetics being used in this are the x and y variables (Income and Life expectantcy), size, and fill color. Income is being mapped to x, life expectancy is being mapped to y, population is mapped to size, and four regions is mapped to fill color. 

2. Import the q1data.csv dataset.

```python
import pandas as pd

#load in data
q1data = pd.read_csv("/Users/chloefeehan/Desktop/python/GSB544/Labs/Lab1/data/q1data.csv")
q1data = q1data.dropna()
```

```python
from plotnine import ggplot, geom_point, aes, labs, scale_size, scale_x_continuous, scale_y_continuous
from plotnine import theme, theme_bw, scale_x_log10, scale_fill_manual, element_text
from plotnine import guides, geom_text, scale_fill_discrete, scale_x_continuous, scale_y_log10


(ggplot(q1data,
aes(x = "income", y = "life_exp", size = "population", fill = "four_regions"))
+ geom_text(aes(x = 7000, y = 60, label="2010"), size=120, color = "lightgray")
+ geom_point(alpha = 0.7, color = "black", stroke = 0.2)
+ labs(x= "Income", y = "Life expectancy", fill = "World Regions")
+ scale_size(range = [.1, 24])
+ scale_x_log10(limits=[0, 128000], 
breaks=[500, 1000, 2000, 4000, 8000, 16000, 32000, 64000],
labels=["500", "1000", "2000", "4000", "8000", "16K", "32K", "64K"])
+ scale_y_continuous(limits = (10, 90), breaks = range(10, 91, 10))
+ scale_fill_manual(values={'africa': 'cyan', 'americas': 'lightgreen', 'asia': 'red', 'europe': 'yellow'},
labels = ["Africa","Americas","Asia","Europe"])
+ theme_bw()
+ theme(text=element_text(family='Comic Sans', size=12),
axis_title=element_text(family='Comic Sans', size=11, color = "darkblue"),
axis_text=element_text(family='Comic Sans', size=8, color = "gray"))
#+ guides(size = "none")
+ theme(legend_position="none")
#+ theme(plot_margin = {t=0.5, r=0.5, b=0.5, l=0.5}, units='cm',)

)
```
