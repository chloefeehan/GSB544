```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
response = requests.get("https://en.wikipedia.org/wiki/List_of_United_States_cities_by_population")
soup = BeautifulSoup(response.content, "html.parser")
```

```python
# find tables 
tables = soup.find_all("table", attrs = {
    "class": "wikitable",
    "style": "text-align:right"})

# see how many tables there are with these attributes
len(tables)

# index table 0, that is the one we want
table = tables[0]
```

```python
# empty list
rows = []

# find every table row
for tr in table.find_all("tr")[1:]: 

    # get all data in columns
    cells = tr.find_all("td")
    
    if len(cells) !=0:
        # extract city name
        city = cells[0].find("a").get_text(strip = True)

        # extract state abbreviation
        state = cells[1].find("a").get_text(strip = True)

        # extract population
        population = cells[2].get_text(strip = True)
        
        # extract land area
        land = cells[5].get_text(strip = True)
    
        rows.append({
            "City": city,
            "State": state,
            "Population": population,
            "Land Area" : land
        })

# convert to df       
df = pd.DataFrame(rows)
df.head()
```

```python
response = requests.get("https://catalog.calpoly.edu/collegesandprograms/collegeofsciencemathematics/statistics/#courseinventory")
soup = BeautifulSoup(response.content, "html.parser")
```

```python
courses = soup.find_all("div", {"class": "courseblock"})

course_list = []

for course in courses:
    name = course.find("strong").get_text(strip = True)

    #get course num by splitting
    course_num = name.split(".", 1)[0].strip()
    course_title = name.split(".", 2)[1].strip()
    course_units = name.split(".", 2)[2].strip()
    
    term = course.find("p", class_ = "noindent").get_text(strip = True)
    term_offered = term.split(":")[1].strip()

    course_list.append({
        "Course Number" : course_num,
        "Course Name" : course_title,
        "Term Offered" : term_offered,
        "Units" : course_units
    })

# convert to df
df_courses = pd.DataFrame(course_list)
df_courses

```
