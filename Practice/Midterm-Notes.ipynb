## Function Documentation

```python
def rescale_vector(vec):
  
  """
  Rescales a numeric vector to have a max of 1 and min of 0
  
  Parameter
  ---------
  vec : num
    A list, numpy array, or pandas Series of numeric data.

  Returns
  -------
  array 
    A numpy array containing the rescaled values.
  """
  
  min_vec = min(vec)
  max_vec = max(vec)
  
  rescaled_vec = (vec - min_vec)/(max_vec - min_vec)
  
  return rescaled_vec
```

### Web Scraping using BeautifulSoup

```python
import requests
from bs4 import BeautifulSoup
response = requests.get("https://statistics.calpoly.edu/content/directory")
soup = BeautifulSoup(response.content, "html.parser")
```

```python
results = soup.find(id="Faculty")
print(results.prettify())
```

```python
# find elements by html class name - need to put in inputs
results.find_all("element name", class_="class name")
```

```python
# find elements by tag
tables = soup.find_all("table")
len(tables)
```

```python
table = tables[1]
table
```

```python
# webscraping faculty data

# initialize an empty list
rows = []

# iterate over all rows in the faculty table
for faculty in table.find_all("tr")[1:]:

    # Get all the cells (<td>) in the row.
    cells = faculty.find_all("td")

    # The information we need is the text between tags.

    # Find the the name of the faculty in cell[0]
    # which for most faculty is contained in the <strong> tag
    name_tag = cells[0].find("strong") or cells[0]
    name = name_tag.text

    # Find the office of the faculty in cell[1]
    # which for most faculty is contained in the <a> tag
    link = cells[1].find("a") or cells[1]
    office = link.text

    # Find the email of the faculty in cell[3]
    # which for most faculty is contained in the <a> tag
    email_tag = cells[3].find("a") or cells[3]
    email = email_tag.text

    # Append this data.
    rows.append({
        "name": name,
        "office": office,
        "email": email
    })
```

## Regular Expressions

```python
moses_string = "Moses supposes his toeses are roses, but Moses supposes erroneously.  Moses he knowses his toeses aren't roses, as Moses supposes his toeses to be."
```

```python
import re

# finds everything that starts with either an "M" or "r" and ends with -oses
re.findall(r"[Mr]oses", moses_string)
```

```python
re.findall(r"\woses", moses_string)
```

```python

# Sample text for regex matching
text = "Hello, World! 1234 is a number. Let's test regex: Python3, regex, and code."

# 1. \b - Matches the boundary between a word and a non-word
print("Words that start with 'r':", re.findall(r'\br\w+', text))

# 2. \s - Matches whitespace characters
print("Whitespace characters:", re.findall(r'\s', text))

# 3. \d - Matches any single digit
print("Digits in the text:", re.findall(r'\d', text))

# 4. ^ - Matches the start of a string
print("Does the text start with 'Hello'? :", bool(re.match(r'^Hello', text)))

# 5. $ - Matches the end of a string
print("Does the text end with 'code.'? :", bool(re.match(r'code\.$', text)))

# 6. . - Matches any character except a newline
print("Characters after 'test':", re.findall(r'test(.)', text))

# Example of finding all words (using \w)
print("All words in the text:", re.findall(r'\w+', text))
```

## Repetition


```python
names = "Key, Kely, Kelly, Kellly, Kelllly"
re.findall(r"Kel+y", names)
```

```python
# Find all occurrences of 'Kely', 'Kelly', or 'Kel' 
# followed by zero or more 'l's in the string 'names'.
print(re.findall(r"Kel*y", names))

# Find all occurrences of 'Kely' or 'Kelly' where 'l' 
# may occur zero or one time (so only one 'l' is allowed).
print(re.findall(r"Kel?y", names))

# Find all occurrences of 'Kelly' where 'l' 
# occurs exactly two times (must match 'Kel' followed by two 'l's).
print(re.findall(r"Kel{2}y", names))

# Find all occurrences of 'Kely' or 'Kelly' where 'l' occurs 
# two to three times (matches 'Kel' followed by two or three 'l's).
print(re.findall(r"Kel{2,3}y", names))

# Find all occurrences of 'Kely' or 'Kelly' where 'l' 
# occurs two or more times (matches 'Kel' followed by at least two 'l's).
print(re.findall(r"Kel{2,}y", names))
```

```python
# extract words from peter piper example that start with p 
nursery_rhyme = "Peter Piper picked a peck of pickled peppers, A peck of pickled peppers Peter Piper picked; If Peter Piper picked a peck of pickled peppers, Wheres the peck of pickled peppers Peter Piper picked?"

re.findall(r"\b[Pp]\w+", nursery_rhyme)
```
